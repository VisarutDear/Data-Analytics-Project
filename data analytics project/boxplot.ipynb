```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
df = pd.read_csv("videogame-sales.csv")
df.head()
```

```python
fieldNeed = ["Platform", "Genre", "Publisher"]
uniqueDict = dict.fromkeys(fieldNeed, [])
for field in fieldNeed:
    temp = df[field].unique()
    uniqueDict[field] = dict.fromkeys(temp, [])
```

```python
# (uniqueDict[fieldNeed[1]])['Sports']
for index, row in df.iterrows():
    # uniqueDict[fieldNeed[0]][row[fieldNeed[0]]]
    for item in fieldNeed:
        temp = uniqueDict[item][row[item]].copy()
        temp.append(row['Global_Sales'])
        uniqueDict[item][row[item]] = temp
```

```python
for i in uniqueDict:
    for j in uniqueDict[i]:
        fig = plt.figure() 
        plt.title("{}-{}".format(i, j))
        plt.boxplot(uniqueDict[i][j])
        plt.show() 
    break
```
