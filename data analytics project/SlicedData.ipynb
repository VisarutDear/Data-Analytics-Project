```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
filename = "videogame-sales.csv" 
chucks = 10 # How many chunk will the dataset divided to.
column_name = "Global_Sales"

df = pd.read_csv(filename)
g_sales = df[column_name].sort_values()
g_sales.head()
```

```python
temp = np.array_split(g_sales, chucks)
print(temp)
```

```python
for data in temp:
    plt.bar(range(len(data)), data)
    plt.show()
```
