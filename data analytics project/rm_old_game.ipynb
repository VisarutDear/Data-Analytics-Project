```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
filename = "videogame-sales.csv" 
chucks = 10 # How many chunk will the dataset divided to.
column_name = "Year"

df = pd.read_csv(filename)
year = df[column_name].sort_values()
year.head()
```

```python
yearSet = list(set(year.dropna().values.tolist()))
print(len(yearSet))
```

```python
amount = []
for i in yearSet:
    amount.append(sum(year==i))

y_pos = np.arange(len(yearSet))

f, ax = plt.subplots(figsize=(18,5))
plt.bar(y_pos, amount, align='center', alpha=0.5)
plt.xticks(y_pos, yearSet, rotation=45)
plt.ylabel('Amount')
plt.title('Amount of game per year')
ax.legend(fontsize = 14)

plt.show()
```

### According to graph
- We can see that in year 1995 is the first year that the number of game released is more than 200 unique title per year. So, we will consider games that released before year 1994 are old games.
